# Step 2 — Create SQLite Database

This notebook will:

1. Create `data/` folder
2. Create `linux_docs.db` database
3. Create tables:
   - `pages`
   - `chunks`
   - `crawl_history`
4. Display list of tables to verify database creation.

> This notebook uses Python's built-in `sqlite3` module, so no additional installation is required.

```python
from pathlib import Path
import sqlite3

# Relative path
DB_PATH = Path("../data/linux_docs.db")

# Ensure data folder exists
DB_PATH.parent.mkdir(parents=True, exist_ok=True)

print(f"Database will be created at: {DB_PATH}")

conn = sqlite3.connect(DB_PATH)
```

## Create tables

```python
conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS pages (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    url TEXT UNIQUE NOT NULL,
    title TEXT,
    html TEXT,
    markdown TEXT,
    scraped_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    status TEXT
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS chunks (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    page_id INTEGER NOT NULL,
    section TEXT,
    chunk_index INTEGER,
    content TEXT NOT NULL,
    token_count INTEGER,
    FOREIGN KEY(page_id) REFERENCES pages(id)
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS crawl_history (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    url TEXT UNIQUE NOT NULL,
    last_crawled TIMESTAMP,
    status_code INTEGER,
    etag TEXT,
    last_modified TEXT
);
""")

conn.commit()

print("All tables created successfully.")
```

## Verification

```python
cursor.execute("""
SELECT name
FROM sqlite_master
WHERE type='table'
ORDER BY name;
""")

tables = cursor.fetchall()

print("List of tables:")
for (name,) in tables:
    print("-", name)
```

## Brief Explanation

### `pages`
Stores raw scraping results and parsing results.

### `chunks`
Stores results of splitting documents into small pieces for embedding.

### `crawl_history`
Stores crawl information so that in the future we can perform *incremental crawl* without re-downloading unchanged pages.

```python
conn.close()
print("Done ✅")
```
