# Step 1 — Setup Project

Notebook ini membantu menyiapkan environment proyek sebelum mulai membuat crawler.

Yang akan dilakukan:
1. Mengecek versi Python.
2. Mengecek package yang dibutuhkan.
3. Membuat struktur folder proyek.
4. Membuat file `.env` jika belum ada.
5. Menampilkan struktur folder yang dibuat.


```python
import sys

print("Python Version :", sys.version)

assert sys.version_info >= (3,11), "Disarankan menggunakan Python 3.11 atau lebih baru."

print("\n✅ Python siap digunakan")
```

## Cek dependency

```python
import importlib

packages = [
    "requests",
    "bs4",
    "lxml",
    "trafilatura",
    "tqdm",
    "dotenv",
]

for pkg in packages:
    try:
        importlib.import_module(pkg)
        print(f"✅ {pkg}")
    except ModuleNotFoundError:
        print(f"❌ {pkg} (belum terinstall)")
```

## Membuat struktur folder

```python
from pathlib import Path

folders = [
    "data",
    "data/raw",
    "src",
    "notebooks",
]

for folder in folders:
    Path(folder).mkdir(parents=True, exist_ok=True)

print("Folder berhasil dibuat.")
```

## Membuat file `.env`

```python
from pathlib import Path

env_path = Path(".env")

default_env = '''DATABASE_PATH=data/linux_docs.db
USER_AGENT=LinuxDocBot/1.0
REQUEST_DELAY=1
'''

if not env_path.exists():
    env_path.write_text(default_env, encoding="utf-8")
    print(".env berhasil dibuat.")
else:
    print(".env sudah ada, tidak ditimpa.")
```
