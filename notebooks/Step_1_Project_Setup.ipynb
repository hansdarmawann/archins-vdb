# Step 1 — Setup Project

This notebook helps set up the project environment before starting the crawler.

What will be done:
1. Check Python version.
2. Check required packages.
3. Create project folder structure.
4. Create `.env` file if it doesn't exist.
5. Display the created folder structure.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import sys

print("Python Version:", sys.version)

assert sys.version_info >= (3,11), "Python 3.11 or newer is recommended."

print("\n✅ Python is ready to use")
```

## Check dependencies

```python
import importlib

packages = [
    "requests",
    "bs4",
    "lxml",
    "trafilatura",
    "tqdm",
    "dotenv",
]

for pkg in packages:
    try:
        importlib.import_module(pkg)
        print(f"✅ {pkg}")
    except ModuleNotFoundError:
        print(f"❌ {pkg} (not installed)")
```

## Create folder structure

```python
from pathlib import Path

folders = [
    "data",
    "src",
    "notebooks",
]

for folder in folders:
    Path('../'+folder).mkdir(parents=True, exist_ok=True)

print("Folders created successfully.")
```

## Create `.env` file

```python
from pathlib import Path

env_path = Path("..\\.env")

default_env = '''DATABASE_PATH=data/linux_docs.db
USER_AGENT=LinuxDocBot/1.0
REQUEST_DELAY=1
'''

if not env_path.exists():
    env_path.write_text(default_env, encoding="utf-8")
    print(".env file created successfully.")
else:
    print(".env file already exists, not overwritten.")
```
